````python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent / "code" / "chapter_03"))

# build_index() scans every file in a folder and builds an inverted index: a
# dict mapping each word to the set of filenames it appears in. search() then
# just looks up a word in that index instead of re-reading every file.
from keyword_search import build_index, search

TEXT_DIR = Path.cwd().parent / "datasets" / "ddr_text_expanded"
index = build_index(TEXT_DIR)

# An f-string (the f before the quotes) lets you drop a Python expression
# straight into a string using {curly braces} -- here, the token count.
print(f"Indexed {len(index)} unique tokens")
````

````python
# search() returns a Python `set` -- an unordered collection with no
# duplicate entries -- of every filename containing the word "fishing".
results = search(index, "fishing")
print("search('fishing'):")
for name in sorted(results):
    print(" ", name)
````
