````python
import sys
from pathlib import Path

# Make the functions in code/chapter_02/expand_abbreviations.py importable here.
sys.path.append(str(Path.cwd().parent / "code" / "chapter_02"))

# ABBREVIATIONS is the dict of drilling shorthand mapped to plain English
# (e.g. "BHA" -> "bottom hole assembly"). expand_text() is the function that
# applies it to a block of text.
from expand_abbreviations import ABBREVIATIONS, expand_text

# Folder holding the plain-text files produced by Chapter 1's batch extraction.
TEXT_DIR = Path.cwd().parent / "datasets" / "ddr_text"
````

````python
# .read_text() opens the file, reads its entire contents into memory as one
# string, and closes the file automatically -- a shortcut for the more manual
# open() / read() / close() sequence.
raw = (TEXT_DIR / "FORGE-16A-78-32_Drilling_038_2020-11-26.txt").read_text()
expanded = expand_text(raw)

# .splitlines() breaks one big string into a list of individual lines, so we
# can loop over them and check each one separately.
for line in expanded.splitlines():
    if "bottom hole assembly 21" in line:
        print(line)
````

Confirm no unrelated word gets corrupted by checking the survey table lines (which contain `MD` and `TVD` as column headers) still read sensibly after expansion.

````python
# Look for the survey table by its section header, or for the expanded form
# of "MD" ("measured depth"). `.lower()` makes the check case-insensitive so
# "Measured Depth" and "measured depth" both match.
for line in expanded.splitlines():
    if "SURVEY DATA" in line or "measured depth" in line.lower():
        print(line)
````
