The ten curated real DDR PDFs used here (from the public Utah FORGE archive, well FORGE 16A(78)-32) are already in `datasets/sample_ddrs/` — no setup needed.

```python
# Python's `sys.path` is the list of folders it searches when you write `import`.
# The notebook lives in `notebooks/`, but the code we want to reuse lives in
# `code/chapter_01/`, so we add that folder to the search path before importing it.
import sys
from pathlib import Path

# Path.cwd() is this notebook's current folder (`notebooks/`); `.parent` steps
# up one level to the book root, then we build the path down into `code/chapter_01/`.
sys.path.append(str(Path.cwd().parent / "code" / "chapter_01"))

# Import the `extract_text` function we wrote and tested in Chapter 1's main
# script (code/chapter_01/read_ddr.py) -- same function, reused here.
from read_ddr import extract_text

# Folder holding the ten real sample DDR PDFs used throughout Part I.
SAMPLE_DIR = Path.cwd().parent / "datasets" / "sample_ddrs"
```

```python
# Build the full path to one real DDR PDF, then run it through the text
# extractor imported above. `text` is now a single Python string containing
# every page's text, with "--- Page N ---" markers inserted between pages.
pdf_path = SAMPLE_DIR / "FORGE-16A-78-32_Drilling_038_2020-11-26.pdf"
text = extract_text(pdf_path)
print(text)
```
